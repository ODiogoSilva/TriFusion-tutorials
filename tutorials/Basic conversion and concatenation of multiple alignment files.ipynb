# Loading the data

As already covered in a separate tutorial (see [The three ways to load data into TriFusion](http://nbviewer.jupyter.org/github/ODiogoSilva/TriFusion-tutorials/blob/master/tutorials/The%20three%20ways%20to%20load%20data%20into%20TriFusion.ipynb)), alignment data can be loaded into TriFusion in three different ways. Here we will use the filechooser to load an entire directory where 614 alignments files are stored. 

Navigate to _Menu_ > _Open/View Data_ and click the __Open file(s)__ button. This will open the main filechooser.

# Setting up conversion options

The conversion and concatenation options are found in the __Proces__ screen. In this screen, select the __Conversion__ button to reveal the general options. 

<img width="90%" src="https://raw.githubusercontent.com/ODiogoSilva/TriFusion-tutorials/master/tutorials/images/process_conversion.png">

- The first option, _Data set_, specifies which dataset will be used for the conversion operation. For now we'll leave it in the default value.

- In the second option, you can choose one or more output formats to convert the input data. In this case we will choose 4 output formats (Fasta, phylip, nexus and stockholm). Some output formats also contains specific additional options that can be viewed by clicking the corresponding __settings__ button. 

<figure>
    <br>
    <p style="font-size: 14px; text-align: center; font-weight: bold;">Click figure to animate</p>
    <img class="animation" width="90%" src="https://raw.githubusercontent.com/ODiogoSilva/TriFusion-tutorials/master/tutorials/images/process_output_formats.png" alt="Static Image" data-alt="https://raw.githubusercontent.com/ODiogoSilva/TriFusion-tutorials/master/tutorials/gifs/process_tutorial3_select_format.gif">
</figure>

- In the third option, select the directory where the output files will be generated. In the conversion operation, the output file names will be based on the corresponding input file (for instance, the input _alignment.fas_ will be converted into _alignment.nex_). However, you can specify a suffix that will be appended to the end of every output file in the __Suffix__ text box. For example, specifying "<i>_variant1</i>" as the suffix will create output files like "<i>alignment1_variant1.nex</i>". 

<img width="90%" src="https://raw.githubusercontent.com/ODiogoSilva/TriFusion-tutorials/master/tutorials/images/process_output_dir.png">

After setting up these general options, you can click the __View Queue__ button at the bottom of the screen to get an overview of the selected options. There you'll see that the 614 files are set to be converted into 4 output formats in a number of output formats whose name will be based on the input. 

<img width="90%" src="https://raw.githubusercontent.com/ODiogoSilva/TriFusion-tutorials/master/tutorials/images/process_conversion_queue.png">

# Setting up concatenation options

After selecting the __Concatenation__ button, the three revealed general options are similar to the ones of the __Conversion__ operation, with the exception that instead of specifying an output __directory__, you'll need to specify an output __file__.

<img width="90%" src="https://raw.githubusercontent.com/ODiogoSilva/TriFusion-tutorials/master/tutorials/images/process_concatenation.png">

We will set the same values for the _Data set_ and _Output Format_ options as in the __Conversion__ operation. In the _Output File_ option we will need to specify the directory AND name of the output file. Let's name our output file "<i>my_concatenation</i>".

# Execution

The execution of either __Conversion__ or __Concatenation__ operations is started by clicking the __Execute__ button at the bottom of the screen. This will open an execution summary with information on the selected main operation, the selected secondary operations (if any), the selected output formats and the expected number of output files. In the case of __Concatenation__ the actual output file name should appear. 

<img width="90%" src="https://raw.githubusercontent.com/ODiogoSilva/TriFusion-tutorials/master/tutorials/images/process_execution.png">

If you're happy with these settings, click the __Execute__ button, and the __Conversion__/__Concatenation__ operation will be carried out. At the end of the execution, an informative popup should appear with a notification that all files were successfully processed.

# Performing concatenation on custom active data set

In many cases, additional operations may be desired on specific subsets of the _total_ loaded dataset. Here I'll demonstrate one way of performing an additional concatenation operation on a custom made data set. 

## Creating and changing the active data set

Suppose I was interested in concatenating the same 614 files, but only for taxa names that start with the letter "A". And after that for taxa names that start with the letter "C". Since I need to create two taxa groups (say, <i>A_taxa</i> and <i>C_taxa</i>), I will also demonstrate two methods to create these data sets.

### Setting "A_taxa" group using the sidepanel toggling method

To create an active data set that contains only taxa whose names start with an "A", go to _Menu_ > _Open/View Data_ and selected the _Taxa_ tab. There are three taxa whose name starts with an "A". The quickest way to selected only these taxa would be to click the __Deselect All__ button and then toggling ON the desired three taxa.

</figure>

### Creating "C_taxa" group using the data set creation dialog

To create the <i>C_taxa</i> via the data set creation dialog, go to _Menu_ > _Dataset Groups_, click the _Taxa_ tab, and then the __Set new taxa group__ button. Since we're dealing with a small number of taxa, we will set the taxa group manually in TriFusion. In the taxa group creation dialog, select the taxa with names starting with a "C" (here using shift + clicking to selected the seven taxa is convenient), specify the group name and click __OK__.

## Execution with custom active data sets

We'll start with the execution of the __Concatenation__ of the 614 files for the <i>A_taxa</i> taxa group. We need to make sure that the value of the _Data set_ general option is set to _Active taxa_, so that TriFusion will use the three active taxa previously defined. Then, click __Execute__ and complete the concatenation operation as before. 

Now for the <i>C_taxa</i> group, select the name of this group in the drop down menu of the _Data set_ general option.
